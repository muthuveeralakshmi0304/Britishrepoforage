```python
import pandas as pd

# Load the reviews from your CSV
df = pd.read_csv("/content/british_airways_reviews.csv")
df.head()
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return " ".join([word for word in text.split() if word not in stop_words])

df['cleaned_review'] = df['review'].apply(clean_text)
df[['review', 'cleaned_review']].head()
```

```python
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

df['sentiment_score'] = df['cleaned_review'].apply(get_sentiment)

# Categorize sentiment
df['sentiment'] = df['sentiment_score'].apply(
    lambda x: 'positive' if x > 0 else 'negative' if x < 0 else 'neutral'
)

df[['cleaned_review', 'sentiment_score', 'sentiment']].head()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

all_words = " ".join(df['cleaned_review'])
wc = WordCloud(width=1000, height=500, background_color='white').generate(all_words)

plt.figure(figsize=(15, 7))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title("Common Words in Reviews", fontsize=20)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='sentiment', palette='Set2')
plt.title("Sentiment Distribution of British Airways Reviews")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.show()
```
